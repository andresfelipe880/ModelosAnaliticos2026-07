**Ejercicios de las Condicionales**

Crea un programa que verifique si una persona es mayor de edad. Una persona es considerada mayor de edad si tiene 18 años o más.

```python
import IPython

# Personas mayores de edad
personas={
    "Edad":"mayor de 18 años",
    "Numero":"par",
    "niños":"0-12 años",
    "adolencente":"13-17 años",
    "adultos":"18-64 años",
    "anciano":"65 años",
    "mayor":"65 años",
    "menor":"0 años",
    "igual":"30 años"
}

IPython.display.display(personas)
```

```python
# Construccion menu del usuario
print("----Bienvenido a tu programa de verificacion de edad----")
print("para comenzar elige la opcion deseada")
print("1. Agregar edad nueva")
print("2. Actulizar la edad")
print("3. Eliminar un dato con del")
print("4. Eliminar un dato con pop")
print("5. Consultar un dato")
print("6. Imprimir diccionario completo")
Seleccion=int(input("selecciona la opcion deseada:"))
```

```python
# Agregar un dato nuevo
if Seleccion==1:
  clave=input("Ingrese la clave deseada:")
  valor=input("ingreses el valor deseado")

  # Agregar un elemento nuevo
  personas[clave]=valor
  print("el valor se agrego exitosamente")
  print(personas)

  # Actulizar
elif Seleccion==2:
  clave=input("ingrese la clave a actulizar")
  personas[clave]=input("ingrese valor nuevo")
  print("el dato fue actulizado con existo")
  print(personas)

  # Eliminar
elif Seleccion==3:
  clave=input("ingrese la clave a eliminar")
  del personas[clave]
  print("el dato fue eliminado exitosamente")

elif Seleccion==4:
  clave=input("ingrese la clave a eliminar")
  personas.pop(clave)
  print("el dato fue eliminado exitosamente")
  print(personas)

  # Consultar dato
elif Seleccion==5:
  clave=input("ingreses la clave a consultar")
  valor=personas.get(clave,"no disponible")
  print(f"{clave}:{valor}")

elif Seleccion==6:
  print("perfil completo")
  print(personas)

else:
  print("opcion invalida")
```
