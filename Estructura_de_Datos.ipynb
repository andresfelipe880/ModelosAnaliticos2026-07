## Lista

```python
# una lista vacia
mi_lista=[]
# mi lista con varios_elemtos
mi_lista=[1,2,3,"4",5.0,True]
print(mi_lista[2])
```

```python
# cambios en la lista
mi_lista[0]="Lionel"
print(mi_lista)
```

```python
# agregar elementos a una lista
mi_lista.append("Messi")
print(mi_lista)
```

```python
# agregar elementos a una posicion especifica
mi_lista.insert(3,"Argentina")
print(mi_lista)
```

```python
# agregar varios elemento al final de la lista
mi_lista.extend([7,False])
print(mi_lista)
```

```python
# quitar elementos
# eliminar objetos especificos
mi_lista.remove(7)
print(mi_lista)
```

```python
print("ACTULIZACION DE DATOS")
valor_buscar=input("Por favor, ingrese el nombre para actulizar:")
valor_reemplazar=input("ingrese valor a reemplazar:")

# Verificamos si el valor a buscar existe en la lista antes de intentar actualizarlo
if valor_buscar in mi_lista:
    mi_lista[mi_lista.index(valor_buscar)]=valor_reemplazar
    print("cambio ejecutado. imprimiendo resultado")
    print(mi_lista)
else:
    print(f"Error: '{valor_buscar}' no se encuentra en la lista. No se pudo realizar el cambio.")
    print(mi_lista) # Mostramos la lista sin cambios
```
