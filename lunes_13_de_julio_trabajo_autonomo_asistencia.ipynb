# LAS 6 OPCIONES DEL MENÚ (EVALUACIÓN CON CONDICIONALES)
AGREGAR UN DATO NUEVO

```python
if opcion == "1":
  nueva_clave = input("ingresa el nombre  de la nueva clave (direccion): ")
  nuevo_valor = input("ingresa el valor para esa clave: ")
  Perfil[nueva_clave] = nuevo_valor
  print("\n[EXITO] Dato agregado correctamente.")
  print("perfil actulizado:", Perfil)
```

#ACTULIZAR UN DATO EXISTENTE

```python
if opcion == "2":
  clave_actulizar = input("ingresa la clave que deseas actulizar: ")
  if clave_actulizar in Perfil:
    nuevo_valor = input("ingresa el nuevo valor: ")
    Perfil[clave_actulizar] = nuevo_valor
    print("\n[EXITO] Dato actualizado correctamente.")
    print("perfil actualizado:", Perfil)
  else:
    print("\n[ERROR] la clave no existe en el perfil.")
```

##ELIMINAR UN DATO DEL

```python
if opcion == "3":
  clave_eliminar = input("ingresar la clave que quieres eliminar con 'del': ")

  if clave_eliminar in Perfil:
    del Perfil[clave_eliminar ]
    print("\n[EXITO] clave eliminada de forma permanente. ")
    print("Perfil actulizado:", Perfil)
  else:
    print("\n[ERROR] la clave no existe, no se pudo eliminar. ")
```

## ELIMININAR UN DATO CON. POP()

```python
if opcion == "4":
    clave_pop = input("Ingresa la clave que quieres eliminar con '.pop()': ")

    if clave_pop in Perfil:
        valor_eliminado = Perfil.pop(clave_pop)
        print(f"\n[ÉXITO] Se eliminó la clave '{clave_pop}'.")
        print(f"El valor que contenía y que fue guardado es: {valor_eliminado}")
        print("Perfil actualizado:", Perfil)
    else:
        print("\n[ERROR] La clave no existe, no se pudo eliminar.")
```

##CONSULTAR UN DATO CON .GET()

```python
if opcion == "5":
    clave_consultar = input("Ingresa la clave que deseas consultar: ")


    resultado = Perfil.get(clave_consultar, "No disponible")
    print(f"\nResultado de la consulta: {resultado}")
```
