##Ejercicio 1: Creación y Modificación de *Listas*

```python
# una lista vacia
mi_lista=[]
# mi lista con varios_elemtos
mi_lista=[1,2,3,"4",5.6,True]
print(mi_lista[2])
```

```python
# quitar elementos
# eliminar objetos especificos
mi_lista.remove("4")
print(mi_lista)
```
