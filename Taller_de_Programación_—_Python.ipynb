```python
#1.Predice la salida
#¿que imprime el for?
for i in range(5):
    print(i)

# Bloque B
for x in range(2, 10, 3):
    print(x)

# Bloque C
for letra in "Hola":
    print(letra)
```

```python
#2.Escribe codigo
#Suma acumulada

# Inicializamos los acumuladores en 0
suma_total = 0
suma_pares = 0

# Ciclo del 1 al 100 (usamos 101 porque el límite superior es exclusivo)
for i in range(1, 101):
    suma_total += i          # Suma todos los números

    if i % 2 == 0:
        suma_pares += i      # Suma solo si el número es par

# Impresión de los resultados
print("La suma de todos los números del 1 al 100 es:", suma_total)
print("La suma de solo los números pares del 1 al 100 es:", suma_pares)
```

```python
#3. escribe codigo
#recorrer la lista
# Lista de precios inicial
precios = [12000, 45000, 8500, 67000, 3200]

# Variable para llevar la cuenta de los artículos caros
contador_caros = 0

# Ciclo para recorrer la lista
# enumerate(precios, 1) nos da la posición (empezando en 1) y el valor
for i, precio in enumerate(precios, 1):
    # 1. Imprimir cada precio con el formato solicitado
    print(f"Item {i}: ${precio:,}")

    # 2. Verificar si cuesta más de $10,000
    if precio > 10000:
        contador_caros += 1

# Imprimir el total de artículos mayores a $10,000 al final
print(f"\nTotal de artículos que cuestan más de $10,000: {contador_caros}")
```

```python
#4. escribe codigo
#While con condición
# Variables iniciales
saldo = 1000
retiro = 150
comision = 20
mes = 0

# El ciclo se ejecuta mientras el saldo sea mayor a cero
while saldo > 0:
    mes += 1
    # Se descuenta el retiro y la comisión en cada mes
    saldo -= (retiro + comision)

    # Imprimir el saldo al final de cada mes
    print(f"Mes {mes}: Saldo restante = ${saldo}")

print(f"\n¡Simulación terminada! Tomó {mes} meses para que el saldo se agotara.")
```

```python
#5. escribe codigo
#break y continue
# Lista inicial de números
numeros = [3, 7, -2, 15, 4, -8, 9, 1]

# Ciclo para evaluar la lista
for num in numeros:
    # 1. Saltear los números negativos
    if num < 0:
        continue

    # 2. Detener el ciclo si el número es mayor a 10
    if num > 10:
        break

    # 3. Imprimir el número procesado (solo si pasó los filtros anteriores)
    print(num)
```

```python
#5. escribe codigo
#Tabla de multiplicar
# Ciclo externo: controla de qué número es la tabla (del 1 al 5)
for i in range(1, 6):

    # Ciclo interno: controla el multiplicador (del 1 al 10)
    for j in range(1, 11):
        # Imprimimos la operación. Usamos end=" " para que no haga un salto de línea
        print(f"{i}x{j}={i*j}", end=" ")

    # Cuando termina una tabla completa, hacemos un print vacío para saltar de línea
    print()
```
