REGRESION LOGISTICA

DIPLOMADO EN ANALITICA DE DATOS

```python
#1. paso, importar librerias

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#2. paso, importar el set de datos

dataset = pd.read_csv("/content/sample_data/Social_Network_Ads.csv")
```

```python
dataset
```

```python
#3 paso, seperacion de variables dependientes e independientes

X = dataset.iloc[:,:-1].values
y = dataset.iloc[:,-1].values
```

```python
#4 paso, separacion datos de entrenamiento y datos de prueba

from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=0)
```

```python
X_train
```

```python
#5. paso, escalada de caracteristicas

from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.fit_transform(X_test)
```

```python
X_train
```

```python
#6 paso, entrenamiento de regresion lostica

from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(random_state=0)
classifier.fit(X_train,y_train)
```

```python
print(classifier.predict(sc.transform([[30,87000]])))
```

```python
#8. paso, prediccion dataset de text

y_pred = classifier.predict(X_test)
dataset["prediccion"] = y_pred
```

```python
dataset
```

```python
#9 paso, matrix de confusion
from sklearn.metrics import confusion_matrix, accuracy_score
y_pred = classifier.predict(X_test)
cm = confusion_matrix(y_test,y_pred)
print(cm)
```

```python
#10 paso visualizacion de resultados
from matplotlib.colors import ListedColormap
X_set, y_set = sc.inverse_transform(X_train), y_train
X1, X2 = np.meshgrid(np.arange(start = X_set[:, 0].min() - 10, stop = X_set[:, 0].max() + 10, step = 0.25),
                     np.arange(start = X_set[:, 1].min() - 1000, stop = X_set[:, 1].max() + 1000, step = 0.25))
plt.contourf(X1, X2, classifier.predict(sc.transform(np.array([X1.ravel(), X2.ravel()]).T)).reshape(X1.shape),
             alpha = 0.75, cmap = ListedColormap(('red', 'green')))
plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())
for i, j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1], c = ListedColormap(('red', 'green'))(i), label = j)
plt.title('Logistic Regression (Training set)')
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.legend()
plt.show()
```
