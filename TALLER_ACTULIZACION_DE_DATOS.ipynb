TALLER ACTULIZACION DE DATOS



```python
# Diccionario con datos basicos
estudiante={
    "Nombres":"Andres Felipe",
    "Apellidos":"Peña",
    "Ciudad":"Bogota",
    "Hijos":"No",
    "Carrera":"Finanzas"
}
```

```python
# Construccion menu del usuario
print("----Bienvenido a tu programa de actulizacion de dato----")
print("para comenzar elige la opcion deseada")
print("1. Agregar un dato nuevo")
print("2. Actulizar un dato")
print("3. Eliminar un dato con del")
print("4. Eliminar un dato con pop")
print("5. Consultar un dato")
print("6. Imprimir diccionario completo")
Seleccion=int(input("selecciona la opcion deseada:"))
```

```python
# Agregar un dato nuevo
if Seleccion==1:
  clave=input("Ingrese la clave deseada:")
  valor=input("ingreses el valor deseado")

  # Agregar un elemento nuevo
  estudiante[clave]=valor
  print("el valor se agrego exitosamente")
  print(estudiante)

  # Actulizar
elif Seleccion==2:
  clave=input("ingrese la clave a actulizar")
  estudiante[clave]=input("ingrese valor nuevo")
  print("el dato fue actulizado con existo")
  print(estudiante)

  # Eliminar
elif Seleccion==3:
  clave=input("ingrese la clave a eliminar")
  del estudiante[clave]
  print("el dato fue eliminado exitosamente")

elif Seleccion==4:
  clave=input("ingrese la clave a eliminar")
  estudiante.pop(clave)
  print("el dato fue eliminado exitosamente")
  print(estudiante)

  # Consultar dato
elif Seleccion==5:
  clave=input("ingreses la clave a consultar")
  valor=estudiante.get(clave,"no disponible")
  print(f"{clave}:{valor}")

elif Seleccion==6:
  print("perfil completo")
  print(estudiante)

else:
  print("opcion invalida")
```
